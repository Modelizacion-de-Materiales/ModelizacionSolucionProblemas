```python
import numpy as np
import matplotlib.pyplot as plt
from barrat import barra
from ejercicio2_guia1 import plotlistT, plotFs
import pdb
plt.rc('figure',figsize=(10,8))
plt.rc('font', size=20)
plt.rc('text', usetex=True)
```
