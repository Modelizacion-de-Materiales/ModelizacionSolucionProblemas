# Modelo SIR

```python
import numpy as np
import matplotlib.pyplot as plt
import pdb
import DFSOLVE as dflib
from importlib.machinery import SourceFileLoader

plt.rc('figure',figsize=(15,10))
plt.rc('axes', labelsize=20)
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
plt.rc('legend',fontsize=20)
plt.rc('lines',linewidth=5, markersize=10)
```

```python
def funcionv(t, _Y, gamma=0.35, beta=0.1):
    """
    Par치metros
    ==========
    Y = [S, I, R]
    """
    _Y = _Y.ravel()
    _dY = [ 
            -gamma*_Y[1]*_Y[0],
            gamma*_Y[1]*_Y[0]-beta*_Y[1],
            beta*_Y[1]
        ]
    return np.array(_dY).reshape(-1,1)
```

```python
def simulacion(_gamma, _beta, _Y0, _dt):
    _Y = _Y0.copy()
#    _gamma = 0.9 # tasa de contagios
#    _beta = 0.01 # tasa de recuperaciion
    newfunv = lambda t, sir: funcionv(t, sir, gamma = _gamma, beta = _beta)
    for i, _dt in enumerate(_dt):
        _Y = np.hstack((_Y, dflib.paso_RK(_dt, t[i], _Y[:,-1].reshape(-1,1), newfunv)))
    return _Y
```

```python
def make_the_plots(thisY, thisgamma, thisbeta):
    plt.plot(thisY[0,:], label='susceptibles')
    plt.plot(thisY[1,:], label='Infectados')
    plt.plot(thisY[2,:],label='Recuperados')
    plt.ylabel('Poblaci칩n')
    plt.xlabel('tiempo')
    plt.legend(title='gamma = {}, beta = {}'.format(thisgamma, thisbeta))
    plt.annotate('gamma / beta  = {:.2f}'.format(thisgamma/thisbeta), xy=(0.5, 0.9), xycoords='axes fraction', fontsize=18)
    return plt.gcf(), plt.gca()
```

# Par치metros de c치lculo globales

```python
t = np.linspace(0,100, 1000)
dt = np.diff(t)
Y0 = np.array( [[0.9],  [0.1] , [0.0]]) # = [ S0, I0, R0]
```

# soluciones por caso

```python
gamma = 0.35
beta = 0.1

Y10=simulacion(0.35, 0.1, Y0, dt)

fig, ax = make_the_plots(Y10, gamma, beta)
```
