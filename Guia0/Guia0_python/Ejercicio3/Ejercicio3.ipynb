```python
m = 70 # kg
k = 12.5 # kg / seg
g = -10 # m / seg^2
dt = 2 # seg , 1 seg, 0.5
```

$$ \dfrac{dv}{dt} = g - \dfrac{k}{m} v $$

$$ dv = (g - \dfrac{k}{m} v ) dt $$

$$ v_{i+1} = v_i + (g - \dfrac{k}{m} v_i) dt $$

```python
def newv(vi,thedt, thek=k, them=m):
    thenewv =  vi + (g-thek*vi/them)*thedt
    return thenewv
```

```python
def vsolve(dt):
    v = [0]
    t=[0]
    while t[-1] <50:
        t.append(t[-1]+dt)
        v.append(newv(v[-1], dt))
    return t, v
```

```python
tv = {'2s': vsolve(2), '1s': vsolve(1), '0.5s': vsolve(0.5)}
theplots = [plt.plot(*thistv, label=thisdt) for thisdt, thistv in tv.items()]#[0], tv['2s'][1])
plt.legend()
```
